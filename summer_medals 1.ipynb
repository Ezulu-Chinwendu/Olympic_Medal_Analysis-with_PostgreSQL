```python
%%sql -- Most Successful Country Each Year by Medals

SELECT year, country, COUNT(*) AS total_medals
FROM summer_medals
WHERE country in ('USA', 'CHN', 'RUS')
GROUP BY year, country
ORDER BY year, total_medals DESC;
```

```python
%%sql --Top 10 Athletes with Most Medals by USA, Russia, China

SELECT athlete, country, COUNT(*) AS medal_count
FROM summer_medals
WHERE country IN ('USA', 'RUS', 'CHN')
GROUP BY athlete, country
ORDER BY medal_count DESC
LIMIT 10;
```

```python
%%sql -- tOP 10 Sports with the Highest Medal Counts for USA

SELECT 
    country, 
    sport, 
    COUNT(*) AS Medals  
FROM summer_medals
WHERE country='USA'
GROUP BY country, sport
ORDER BY Medals DESC
LIMIT 10;
```

```python
%%sql -- Gender Distribution of Medalists for USA, Russia, and China

SELECT 
    country, 
    gender, 
    COUNT(*) AS Medals
FROM summer_medals
WHERE country IN ('USA', 'RUS', 'CHN')
GROUP BY country, gender
ORDER BY country, gender;
```

```python
%%sql -- Yearly Trends in Medal Wins by Gender for USA Since 2000

SELECT 
    year, 
    gender, 
    COUNT(*) AS medal_count
FROM summer_medals
WHERE country = 'USA' and Year > 1999
GROUP BY year, gender
ORDER BY year, gender;
```

```python
%%sql -- Most Competitive Disciplines by Medals

SELECT discipline, COUNT(*) AS medal_count
FROM summer_medals
GROUP BY discipline
ORDER BY medal_count DESC
LIMIT 10;
```

```python
%%sql -- USA vs Russia: Most Successful Sport in 2012

SELECT 
    year, 
    sport, 
    country, 
    COUNT(*) AS medal_count
FROM summer_medals
WHERE country IN ('USA', 'RUS')  AND Year=2012
GROUP BY year, sport, country
ORDER BY year, medal_count DESC;
```

```python
%%sql -- Year-on-Year Growth in Medals Since 2000

SELECT 
    year, 
    COUNT(*) AS medals_won,
    COUNT(*) - LAG(COUNT(*)) OVER (ORDER BY year) AS growth
FROM summer_medals
WHERE country = 'CHN' AND Year>1999
GROUP BY year
ORDER BY year;
```

```python
%%sql -- Top 10 Countries with Most Medals in 100M

SELECT event, country, COUNT(*) AS medal_count
FROM summer_medals
WHERE event = '100M'
GROUP BY event, country
ORDER BY medal_count DESC
LIMIT 10; 
```

```python
%%sql -- Percentage Contribution of Each Country to Total Medals Since 2000


WITH yearly_totals AS (
    SELECT year, COUNT(*) AS total_medals
    FROM summer_medals
    GROUP BY year
)
SELECT 
    o.year, 
    o.country, 
    COUNT(*) AS country_medals,
    ROUND((COUNT(*)::NUMERIC / y.total_medals) * 100, 2) AS percentage
FROM summer_medals o
JOIN yearly_totals y ON o.year = y.year
WHERE o.country IN ('USA', 'RUS', 'CHN') AND o.year>1999
GROUP BY o.year, o.country, y.total_medals
ORDER BY o.year, o.country;
```
